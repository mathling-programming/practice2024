```python
import collections
from collections import Counter

def frequency_histogram(sample, number):
  with open(sample) as f:
     text = f.read()
  text = text.lower()
  tokens = text.split()
  frequency = collections.Counter(tokens)
  most_frequent_tokens = frequency.most_common(number)
  for i in range(number):
    a = len(most_frequent_tokens[i][0])
    print(most_frequent_tokens[i][0], ' '*(10-a), '|', '#'*most_frequent_tokens[i][1], most_frequent_tokens[i][1])

  return most_frequent_tokens

most_frequent_tokens = frequency_histogram("sample.txt", 100)
```
